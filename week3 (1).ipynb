```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import json
from geopy.geocoders import Nominatim
import requests 
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans
import folium
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
wiki='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
pages=urlopen(wiki)
soup=BeautifulSoup(pages)
```

```python
postcode=[]
borough=[]
neighborhood=[]

table=soup.find("table",{"class","wikitable sortable"})
print(table)
```

```python
for row in table.findAll('tr'):
    cells=row.findAll('td')
    if(len(cells)==3):
        postcode.append(cells[0].find(text=True))
        borough.append(cells[1].find(text=True))
        neighborhood.append(cells[2].find(text=True))
```

```python
print(postcode)
print(borough)

print(neighborhood) # we have to remove \n in it.
```

```python
neighborhood_2=[]
for i in neighborhood:
    i=i.replace('\n','')
    neighborhood_2.append(i)
```

```python
print(postcode)
print(borough)
print(neighborhood_2)
```

```python
# check length
print(len(postcode))
print(len(borough))
print(len(neighborhood_2))
```

```python
data=pd.DataFrame({"postcode":postcode,"borough":borough,"Neighborhood":neighborhood_2})
```
